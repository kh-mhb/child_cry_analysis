As you can see in the input bar, the audio samples are sorted with some label folders but We do not have any file that give us the label of the audio that helps us to predict the audio label as result. We are going to perform some operation to append the label and put in a dataframe as well.

```python
base_dir = "/kaggle/input/infant-cry-audio-corpus/donateacry_corpus"
data = []

for root, _, files in os.walk(base_dir):
    for file in files:
        if file.endswith(".wav"):  
            file_path = os.path.join(root, file)  
            label = os.path.basename(root) 
            data.append([file_path, label])

df = pd.DataFrame(data, columns=["audio_link", "label"])

csv_path = "/kaggle/working/infant_cry_data.csv"
df.to_csv(csv_path, index=False)

print(f"CSV file saved to: {csv_path}")
```

    <h2 style="text-align: center; color: #d4a373;">New Feature Generation</h2>
    <p style="border-left: 3px solid #d4a373; padding-left: 10px;">
        After the data preperation step, we got two feature: 1. audio_link, 2. label.
        Now we are gonna extract new features from the audio_link, process it by using librosa and create new features like mfccs, spectral centroid, zero crossing rate, chroma.
    </p>
    <ul>
        <li>MFCCs (Mel-Frequency Cepstral Coefficients): Features representing the short-term power spectrum of a sound, mimicking human auditory perception.</li>
        <li>Spectral Centroid: Indicates the center of mass of a spectrum, perceived as the brightness of a sound.</li>
        <li>Zero Crossing Rate (ZCR): Measures the rate at which a signal changes its sign, useful for distinguishing percussive from tonal sounds.</li>
        <li>Chroma Features: Represent the tonal content of audio by mapping frequencies to their corresponding musical pitch classes.</li>
    </ul>
</div>

```python
def process_audio_csv(input_csv, output_csv, n_mfcc=13):
    try:
        df = pd.read_csv(input_csv)
        if "audio_link" not in df.columns:
            print("Error: 'audio_link' column not found in CSV!")
            return
        feature_columns = [f"MFCC_{i+1}" for i in range(n_mfcc)] + \
                          ["Spectral_Centroid", "Zero_Crossing_Rate"] + \
                          [f"Chroma_{i+1}" for i in range(12)]
        for col in feature_columns:
            df[col] = None
        for i, row in df.iterrows():
            audio_path = row["audio_link"]
            try:
                y, sr = librosa.load(audio_path, sr=None)
                mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc), axis=1)
                spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr), axis=1)
                zcr = np.mean(librosa.feature.zero_crossing_rate(y), axis=1)
                chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr), axis=1)
                features = np.concatenate([
                    mfccs, spectral_centroid.ravel(), zcr.ravel(), chroma
                ])
                df.loc[i, feature_columns] = features

                print(f"Processed: {audio_path}")

            except Exception as e:
                print(f"Error processing {audio_path}: {e}")
        df.to_csv(output_csv, index=False)
        print(f"Feature extraction complete. Saved to {output_csv}")
    except Exception as e:
        print(f"Error: {e}")
process_audio_csv("/kaggle/working/infant_cry_data.csv", "final.csv")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation="relu"),
    layers.Dense(16, activation="relu"),
    layers.Dense(len(y.unique()), activation="softmax")
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

loss, acc = model.evaluate(X_test, y_test)
print("Deep Learning Model Accuracy:", acc)
```

```python
def extract_audio_features(audio_path, n_mfcc=13):
    try:
        y, sr = librosa.load(audio_path, sr=None)
        mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc), axis=1)
        spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr), axis=1)
        zcr = np.mean(librosa.feature.zero_crossing_rate(y), axis=1)
        chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr), axis=1)
        features = np.concatenate([
            mfccs, spectral_centroid.ravel(), zcr.ravel(), chroma
        ])
        return features
    except Exception as e:
        print(f"Error extracting features from {audio_path}: {e}")
        return None
```
